```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
plt.plot([1,2,3,4,5],[1,4,9,16,25])
plt.show()
```

```python
#adding details to the figure
plt.plot([1,2,3,4,5],[1,4,9,16,25])
plt.figure(figsize=(8,6))
plt.title("My first plot")
plt.xlabel("numbers")
plt.ylabel("squares")
plt.show()
```

```python
#plotting multiple sets of data
x=np.arange(1,5)
y=x**3
plt.plot([1,2,3,4],[1,4,9,16],'g-',x,y,'r-',[1,2,3,4],[1,2,3,4],'b-')
plt.title("squares and cubes")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.show()
```

```python
plt.subplot(1,3,1)
plt.plot([1,2,3,4],[1,4,9,16],"b-")
plt.title("ist subplot - squares")
plt.subplot(1,3,2)
plt.plot(x,y,'r-')
plt.title("2nd subplot")
plt.subplot(1,3,3)
plt.plot([1,2,3,4],[1,2,3,4],"g-")
plt.suptitle("My Subplots")
plt.show()
```

```python
plt.subplot(2,1,1)
plt.plot([1,2,3,4],[1,4,9,16],'g-')
plt.title("ist subplot")
plt.subplot(2,1,2)
plt.plot(x,y,"r-")
plt.title("2nd subplot")
plt.suptitle("my subplot")
plt.show()
```

```python
x=np.arange(1,5)
y = x**3
fig,ax = plt.subplots(nrows =2, ncols = 2, figsize = (10,10))
ax[0,0].plot([1,2,3,4],[1,4,6,8],'g-')
ax[0,1].plot([1,2,3,4],[1,4,9,16],'ro')
ax[1,0].plot(x,y,'b^')

ax[0,0].set_title("Double the number")
ax[0,1].set_title("squares")
ax[1,0].set_title("cubes")

plt.show()
```

```python
divisions=["Div-A","Div-B","Div-C","Div-D","Div-E","Div-F"]
division_average_marks = [70,82, 98,10,65,68]
plt.bar(divisions,division_average_marks,color= 'lawngreen')
plt.title("Bar Graph")
plt.xlabel("Divisions")
plt.ylabel("Marks")
plt.show()
```

```python
labels = ["chips","pastry", "chocolates", "juice"]
sales = [120,76,350,10]
plt.bar(labels,sales,color= 'aqua')
plt.title("Bar Graph")
plt.ylabel("Products")
plt.xlabel("Qty Sold")
plt.show()
```

```python
months=["Jan","Feb","Mar"]
chips = [10,5,15]
juice = [21,15,45]
candy = [25,28,32]

index = np.arange(3)
width = 0.1
plt.bar(index,chips, width, color='darkorange', label = "CHIPS")
plt.bar(index+width,juice, width, color = 'plum', label ="JUICE")
plt.bar(index+width*2,candy, width, color ='crimson', label = "CANDY")

plt.title("Sales of products")

plt.xlabel("Products")
plt.ylabel("Sales")
plt.xticks(index+width/2, months)
plt.legend(loc='best')
plt.show()
```

```python
#pie charts
category = ["Books","Food","Cosmetics","Travel","Bills","Internet"]
share = [3000,5000,2000,1500,4500, 1200]
plt.pie(share, labels= category, shadow = False)
plt.axis('equal')
plt.show()
```

```python
x = np.random.randn(5000)
plt.title ("Histogram")
plt.xlabel("Random data")
plt.ylabel("Frequency")
plt.hist(x,1000, color ='midnightblue')
plt.show()
```
